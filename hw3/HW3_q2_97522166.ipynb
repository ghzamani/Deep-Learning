```python
import numpy as np
import matplotlib.pyplot as plt

d1 = np.full((1, 100), 1)
d2 = np.full((1, 100), -1)
d = np.concatenate((d1, d2), axis=1)
print("vector d", d)

sum = np.cumsum(d)
avg = [sum[i]/(i+1) for i in range(sum.size)]
print("average of vector d", avg)

plt.plot(d[0], linewidth=4.0)
plt.plot(avg, 'r')
plt.show()
```

```python
beta = 0.9
exp_avg = np.zeros((1, 200))
# let m1[0] = 0 + (1-beta) * d[0]
exp_avg[0][0] = (1-beta) * d[0][0]
for i in range(d.size-1):
  exp_avg[0][i+1] = beta * exp_avg[0][i] + (1-beta) * d[0][i+1]
  
print("exponential moving average of vector d", exp_avg)
plt.plot(d[0], linewidth=4.0)
plt.plot(exp_avg[0], 'r')
plt.show()
```

```python
time = np.arange(0, d.size)
bias_correction = np.zeros((1, 200))
for i in range(1, d.size):
  bias_correction[0][i] = exp_avg[0][i] / (1 - beta**time[i])
  
print("exp avg after bias correction", bias_correction)
plt.plot(d[0], linewidth=4.0)
plt.plot(bias_correction[0], 'r')
plt.show()
```

```python
beta2 = 0.999
exp_avg2 = np.zeros((1, 200))
# let m1[0] = 0 + (1-beta2) * d[0]
exp_avg2[0][0] = (1-beta2) * d[0][0]
for i in range(d.size-1):
  exp_avg2[0][i+1] = beta2 * exp_avg2[0][i] + (1-beta2) * d[0][i+1]
  
print("exponential moving average, beta=0.999", exp_avg2)
plt.plot(d[0], linewidth=4.0)
plt.plot(exp_avg2[0], 'r')
plt.show()
```

```python
bias_correction2 = np.zeros((1, 200))
for i in range(1, d.size):
  bias_correction2[0][i] = exp_avg2[0][i] / (1 - beta2**time[i])
  
print("exp avg after bias correction, beta=0.999", bias_correction2)
plt.plot(d[0], linewidth=4.0)
plt.plot(bias_correction2[0], 'r')
plt.show()
```
