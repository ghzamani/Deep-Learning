```python
import numpy as np
import random

data = {}
data['Iris-setosa'] = []
data['Iris-versicolor'] = []
data['Iris-virginica'] = []

test = []
# process the data

with open('iris.data') as infile:
  for line in infile:
    num = line.split(',')
    # last line shouldn't be read
    if(len(num) == 1):
      break

    label = num[4].strip()
    features = np.array([float(x) for x in num[:4]])
    
    data[label].append(features)
    test.append((features, label))

data['Iris-setosa'] = np.array(data['Iris-setosa'])
data['Iris-versicolor'] = np.array(data['Iris-versicolor'])
data['Iris-virginica'] = np.array(data['Iris-virginica'])

print("class 1 train shape", data['Iris-setosa'].shape)
print("class 2 train shape", data['Iris-versicolor'].shape)
print("class 3 train shape", data['Iris-virginica'].shape)
```

```python
# compute mean and variance for each feature of each class
def compute_mean(v):
  return np.mean(v, axis=0)

def compute_var(v):
  return np.var(v, axis=0)

mean = {}
var = {}
for k in data.keys():
  mean[k] = compute_mean(data[k])
  var[k] = compute_var(data[k])
  print("mean in class", k, mean[k])
  print("var in class", k, var[k])


# compute probability of each label
p = {}
total = len(data['Iris-setosa']) + len(data['Iris-versicolor']) + len(data['Iris-virginica'])
for k in data.keys():
  p[k] = len(data[k]) / total

print(p)
```

```python
def compute_conditional(v, mean, var):
  return (1 / np.sqrt(2 * np.pi * var)) * np.exp((-1 * (v - mean)**2) / (2 * var))

def predict(l, v):
  conditional = compute_conditional(v, mean[l], var[l])
  return p[l] * np.prod(conditional)

# shuffle test data
random.shuffle(test)

print("prediction***************true label")
accuracy = 0
for t in test:
  max_p = 0
  for k in data.keys():
    pr = predict(k, t[0])
    if pr > max_p:
      max_p = pr
      prediction = k
  if prediction == t[1]:
    accuracy += 1

  print(prediction, "**********", t[1])

print("total accuracy is", accuracy/total)
```
