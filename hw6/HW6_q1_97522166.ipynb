```python
import matplotlib.pyplot as plt
import numpy as np
import math
  
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def swish(x):
  return x * sigmoid(x)

x = np.linspace(-6, 6, 100)
y = swish(x)
  
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("Swish(X)")
  
plt.show()
```

```python
def softplus(x):
  return np.log(1 + np.exp(x))

def mish(x):
  return x * np.tanh(softplus(x))

# x2 = np.linspace(-3, 3, 100)
y2 = mish(x)
  
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("Mish(X)")
  
plt.show()
```

```python
y3 = swish(x) + sigmoid(x) * (1 - swish(x))
  
plt.plot(x, y3)
plt.xlabel("x")
plt.ylabel("derivative of Swish(X)")
  
plt.show()
```

```python
y4 = mish(x) / x + x * (1 - np.tanh(softplus(x))**2) * sigmoid(x)

plt.plot(x, y4)
plt.xlabel("x")
plt.ylabel("derivative of Mish(X)")
  
plt.show()
```
